# What was the best month for sales ? How much was earned that month?

```python
df_filtered.info()
```

```python
df_filtered.head()
```

```python
df_filtered["Quantity Ordered"]=df_filtered["Quantity Ordered"].astype("float")
df_filtered["Price Each"]=df_filtered["Price Each"].astype("float")      
```

```python
df_filtered["Price Each"].nunique
```

```python
df_filtered.head()
```

```python
df_filtered["sales"]=df_filtered["Quantity Ordered"]*df_filtered["Price Each"]
```

```python
df_filtered["month"].unique()
```

```python
df=df_filtered.groupby(['month'])['Sales'].sum().sort_values(ascending=False)
```

```python
df=df.to_frame()
```

```python
df.info()
```

```python
df['Sales']
```

```python
df.columns
```

```python
month=["dec","oct","apr","nov","may","mar","july","june","aug",'feb',"sep","jan"] 
df["month"]=month
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.barplot(x = "Sales",
            y = "month",
            data = df)
plt.title("Month wise Sale")
plt.show()
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)


import seaborn as sns
sns.barplot(x = "Sales",
            y = "month",
            data = df)
plt.title("Month wise Sale")
plt.show()
```

We can clearly see that month 12 (December) is the highest sales in 2019 with approximately $4,810,000.

# Which city had the highest number of sales ?

```python
dftemp = df_filtered
list_city = []
for i in dftemp['Purchase Address']:
    list_city.append(i.split(",")[1])
dftemp['City'] = list_city
dftemp.head()
```

you can see that we added city column by extracting city from purchased address

```python
df_city=df_filtered.groupby(["City"])['sales'].sum().sort_values(ascending=False)
df_city=df_city.to_frame()
df_city
```

```python
df_city.head()
```

```python
city=["San Francisco","Los Angeles","New York City","Boston","Atlanta","Dallas","Seattle","Portland","Austin"]
df_city["city"]=city
```

```python
from matplotlib import pyplot
a4_dims = (11.7, 8.27)
fig, ax = pyplot.subplots(figsize=a4_dims)
sns.barplot(x = "Sales",
            y = "city",
            data = df_city)
plt.title("City wise Sales")
plt.show()
```

Sanfrancisco has highest no of sales
